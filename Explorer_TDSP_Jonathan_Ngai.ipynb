**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/Explore TDSP/Explore TDSP Data/Motor_Vehicle_Collisions_Crashes.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=pd.Series(average_crashes_per_hour.index), y=average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes.values, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Time')
plt.ylabel('Number of Collisions')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasons')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Model Construction**: Now, we will create a Logistics Regression Model to predict whether or not a accident will lead to injuries and/or death.

````python
def feature_engine_inital(data, is_test_set=False, enc=None):
    if not is_test_set:
        #processing for the training set
        #training set can involve references to response variable 'Unscathed')
        #can remove data points for training data
        copy = data.copy()
        copy['BOROUGH'] = copy['BOROUGH'].fillna('Unspecified')
        # cleaned_vehicle_data = standard_cleaner(cleaned_borough_data, 'CONTRIBUTING FACTOR VEHICLE 1', 'Unspecified') no longer necessary

        #initialize the encoder only once for training set
        if enc is None:
            enc = OneHotEncoder()  #avoiding dummy variable trap by dropping one column

        #fit the encoder on the training data
        enc.fit(copy[['BOROUGH']])
        encoded_data = enc.transform(copy[['BOROUGH']]).toarray()
        encoded_data_df = pd.DataFrame(encoded_data, columns=enc.get_feature_names_out(['BOROUGH']))

        #combine the encoded columns with the original data, removing 'BOROUGH'
        data_w_ohe = copy[['Hour of Day']].reset_index().join(encoded_data_df)

        X = data_w_ohe
        Y = np.array(copy['Unscathed'])


        return X, Y, enc

    else:
        if enc is None:
            raise ValueError("An initialized encoder must be passed when processing the test set.")

        #processing for the test set
        #test set should not involve the response variable 'Unscathed' or removing rows, model must operate on data independent of response
        #cannot remove data points for test set, model must make a prediction for every data point as if it were functioning in the real world
        copy = data.copy()
        copy['BOROUGH'] = copy['BOROUGH'].fillna('Unspecified')

        encoded_data = enc.transform(copy[['BOROUGH']]).toarray()
        encoded_data_df = pd.DataFrame(encoded_data, columns=enc.get_feature_names_out(['BOROUGH']))

        #combine the encoded columns with the test data, removing 'BOROUGH'
        data_w_ohe = copy[['Hour of Day']].reset_index().join(encoded_data_df)

        #return only predictors (X) for the test set
        X = data_w_ohe
        return X
````

**Poisson Process:** Could we perhaps model the arrival of crashes via the poisson process? For now let's restrict our analysis to 2024.

````python
#we will randomly select a day then see the distribution of gaps
import random

random.seed(42)
random_date = pd.Timestamp('2024-01-01') + pd.to_timedelta(random.randint(0, 364), unit='D')
sample = data[data['CRASH DATE'] == random_date]

sample['Seconds Since Midnight'] = data['CRASH TIME'].dt.hour * 60**2 + data['CRASH TIME'].dt.minute * 60 + data['CRASH TIME'].dt.second
sample = sample.sort_values(by='Seconds Since Midnight')
sample['Gap'] = sample['Seconds Since Midnight'].diff()
sample.loc[sample.index[0], 'Gap'] = 0
sample[['CRASH TIME','Seconds Since Midnight', 'Gap']]
````

````python
from scipy import stats
from scipy.stats import poisson
from sklearn.metrics import mean_squared_error
````

````python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 7))

lam_test = 1 / np.mean(sample['Gap'])
t = np.arange(5000)
f = stats.expon.pdf(t, scale = 1/lam_test)

sns.histplot(data=sample, x='Gap', stat='density') #looks like an exponential distribution, lets look at some more days
plt.xlim(0, 5000)
plt.plot(t, f, color='red', lw=1)

plt.title('Distribution of Gap Sizes in a Sample of a Day from 2024', fontsize=16)
plt.xlabel('Gap (Seconds)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

# Loop through each day and plot the distribution
for i in range(10):  # Here we assume you want 10 days
    current_day = pd.Timestamp('2024-01-01') + pd.to_timedelta(random.randint(0, 364), unit='D')
    s = data[data['CRASH DATE'] == current_day]

    #calculate the gap between crashes in seconds
    Seconds_Since_Midnight = (s['CRASH TIME'].dt.hour * 60**2 + s['CRASH TIME'].dt.minute * 60 + s['CRASH TIME'].dt.second).sort_values()
    Gap = Seconds_Since_Midnight.diff().fillna(0)  # Calculate gap, and fill first NaN with 0

    #calculate the rate for the Poisson distribution
    lam = 1 / np.mean(Gap) if np.mean(Gap) != 0 else 1  # Prevent division by zero

    #set up the range and axies for the probability distribution plots
    t = np.arange(5000)
    f = stats.expon.pdf(t, scale=1/lam)

    #do the actual plotting
    sns.histplot(Gap.values, kde=False, stat='density', ax=axes[i])

    #add the exponential probability density function
    axes[i].plot(t, f, color='red', lw=1)

    # Set axis limits and labels
    axes[i].set_xlim(0, 5000)
    axes[i].set_title(f"Day {current_day.strftime('%Y-%m-%d')}")
    axes[i].set_xlabel('Gap (Seconds)')
    axes[i].set_ylabel('Density (%)')

# Adjust layout
plt.tight_layout()
plt.show()
````

````python
#calculating lambda involving all entries in 2024
data2024 = data[data['CRASH DATE'].dt.year == 2024].sort_values(by=['CRASH DATE', 'CRASH TIME'])
Sec_Since_Midnight = data2024['CRASH TIME'].dt.hour * 60**2 + data2024['CRASH TIME'].dt.minute * 60 + data2024['CRASH TIME'].dt.second
data2024['Time Since First Crash'] = (data2024['CRASH DATE'] - data2024['CRASH DATE'].min()).dt.days * 24*60**2 + Sec_Since_Midnight
data2024['Gap'] = data2024['Time Since First Crash'].diff()
data2024.iloc[0, len(data2024.columns) - 1] = 0.0
````

````python
lam_final = 1 / np.mean(data2024['Gap'])
t = np.arange(2000)
f = stats.expon.pdf(t, scale = 1/lam_final)

sns.set_style("whitegrid")
plt.figure(figsize=(12, 7))

sns.histplot(data=data2024, x='Gap', stat='density', bins= 100)
plt.xlim(0, 2000)
plt.plot(t, f, color='red', lw=1)

plt.title('Distribution of Gap Sizes Across All Entries in 2024', fontsize=16)
plt.xlabel('Gap (Seconds)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
#alternate plot that's more concerning
sns.set_style("whitegrid")
plt.figure(figsize=(12, 7))

sns.histplot(data=data2024, x='Gap', stat='density')
plt.xlim(0, 2000)
plt.plot(t, f, color='red', lw=1)

plt.title('Distribution of Gap Sizes Across All Entries in 2024 with Granular Bin Size', fontsize=16)
plt.xlabel('Gap (Seconds)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

So some caveats: The poisson model **seems reasonable** as most histograms plotting the distribution of gap sizes seem to follow an exponential distribution. Even when plot all of 2024's gaps, it seems gaps tend to follow a exponential distribution.

However, **adjustments to bin size change the distribution of gap size to fit the exponential distribution less.** This is not unsual, having less gaps will smooth over important characteristics in the data whereas having more gaps will create bins with no values and amplify the influence of random noise. In particular in the plot of gap sizes across all of 2024's accidents with more granular bins there seems to be some regular peaking in density every 5-6 bins as seen above that's particularly suspicious. Thus **more research should be conducted to assess the viability of using a poisson process model to predict the number of crashes.**



````python
conversion_factor = 60**2*24*365 #poisson mew for number of accidents per year
annual_expected_crashes = lam_final * conversion_factor
daily_expected_crashes = annual_expected_crashes / 365
hourly_expected_crashes = annual_expected_crashes / (24*365)
actual_num_crashes_2024 = len(data[data['CRASH DATE'].dt.year == 2024])

print(f"Expected number of crashes per year: {annual_expected_crashes}")
print(f"Expected number of crashes per day: {daily_expected_crashes}")
print(f"Expected number of crashes per hour: {hourly_expected_crashes}")
print(f"Actual number of crashes in 2024: {actual_num_crashes_2024}")
````

````python
#normal approximation histogram
normal_approx = np.random.normal(loc=annual_expected_crashes, scale=np.sqrt(annual_expected_crashes), size=10000)

sns.set_style("whitegrid")
plt.figure(figsize=(12, 7))

plt.hist(normal_approx, bins=50, density=True, alpha=0.6, color='g')

plt.axvline(annual_expected_crashes - np.sqrt(annual_expected_crashes), color='blue', linestyle='dashed', linewidth=2, label=f"Mean - 1 SD ({annual_expected_crashes - np.sqrt(annual_expected_crashes)})")
plt.axvline(annual_expected_crashes + np.sqrt(annual_expected_crashes), color='blue', linestyle='dashed', linewidth=2, label=f"Mean + 1 SD ({annual_expected_crashes + np.sqrt(annual_expected_crashes)})")

ci_lower = annual_expected_crashes - 1.96 * np.sqrt(annual_expected_crashes)
ci_upper = annual_expected_crashes + 1.96 * np.sqrt(annual_expected_crashes)

plt.axvline(ci_lower, color='red', linestyle='dashed', linewidth=2, label=f"95% CI Lower Bound ({ci_lower})")
plt.axvline(ci_upper, color='red', linestyle='dashed', linewidth=2, label=f"95% CI Upper Bound ({ci_upper})")

plt.title(f'Histogram of Normal Approx of General NYC Poisson Model (lambda = {annual_expected_crashes})', fontsize=16)
plt.xlabel('Counts', fontsize=14)
plt.ylabel('Density (probability per unit count)', fontsize=14)
plt.tight_layout()
plt.legend()
plt.show()

print(f"95% confidence interval : {(ci_lower, ci_upper)}")
````

````python
name_list = ['QUEENS', 'BROOKLYN', 'BRONX', 'MANHATTAN', 'STATEN ISLAND']
expected_crashes = []
actual_crashes = []
ci = []

for i in range(5):
  name = name_list[i]
  p = probability_vector[i]

  annual_expected_crashes = lam_final * p * conversion_factor
  expected_crashes.append(annual_expected_crashes)

  daily_expected_crashes = annual_expected_crashes / 365
  hourly_expected_crashes = annual_expected_crashes / (24*365)
  actual_num_crashes_2024 = len(data2024[data2024['BOROUGH'] == name])
  actual_crashes.append(actual_num_crashes_2024)

  print(f"""======================================================{name} DATA======================================================\n""")

  nor = np.random.normal(loc=annual_expected_crashes, scale=np.sqrt(annual_expected_crashes), size=10000)

  sns.set_style("whitegrid")

  plt.hist(nor, bins=50, density=True, alpha=0.6)

  plt.axvline(annual_expected_crashes - np.sqrt(annual_expected_crashes), color='blue', linestyle='dashed', linewidth=2, label=f"Mean - 1 SD ({annual_expected_crashes - np.sqrt(annual_expected_crashes)})")
  plt.axvline(annual_expected_crashes + np.sqrt(annual_expected_crashes), color='blue', linestyle='dashed', linewidth=2, label=f"Mean + 1 SD ({annual_expected_crashes + np.sqrt(annual_expected_crashes)})")

  local_ci_lower = annual_expected_crashes - 1.96 * np.sqrt(annual_expected_crashes)
  local_ci_upper = annual_expected_crashes + 1.96 * np.sqrt(annual_expected_crashes)

  plt.axvline(local_ci_lower, color='red', linestyle='dashed', linewidth=2, label=f"95% CI Lower Bound ({local_ci_lower})")
  plt.axvline(local_ci_upper, color='red', linestyle='dashed', linewidth=2, label=f"95% CI Upper Bound ({local_ci_upper})")

  plt.title(f'Histogram of Normal Approx of {name} Poisson Model (lambda = {annual_expected_crashes})')
  plt.xlabel('Counts')
  plt.ylabel('Density (probability per unit count)')
  plt.tight_layout()
  plt.legend()
  plt.show()

  ci.append((local_ci_lower, local_ci_upper))

  print(f"95% confidence interval for expected number of crashes per year in {name}: {(local_ci_lower, local_ci_upper)}")
  print(f"Actual number of crashes in 2024 in {name}: {actual_num_crashes_2024}")
  print(f"Proportion of accidents in 2024 that happened in {name}\n(used as probability for multitype poisson process): {p}\n")


print(f"MSE incorporating 'Unspecified': {mean_squared_error(actual_crashes, expected_crashes)}")
print(f"MSE not incorporating 'Unspecified': {mean_squared_error(actual_crashes[:-1], expected_crashes[:-1])}\n")

print("NOTE: category 'Unspecified' not plotted")
````
